# Dealing with categorical features:

## 1 - Temporal Features

```python
df_Temp = df.copy()
df_Temp['transaction_time'] = pd.to_datetime(df_Temp['transaction_time'])
df_Temp['hour'] = df_Temp['transaction_time'].dt.hour
df_Temp['day_of_week'] = df_Temp['transaction_time'].dt.dayofweek
df_Temp['is_weekend'] = df_Temp['day_of_week'].isin([5, 6]).astype(int)
df_Temp['is_night'] = df_Temp['hour'].between(22, 6).astype(int)  # 10pm-6am
df_Temp['day_of_month'] = df_Temp['transaction_time'].dt.day

df_Temp.head()
```

## 2- Behavioral Features:

```python
df_behave = df_Temp.copy()

df_behave = df_behave.sort_values(['user_id', 'transaction_time'])

# Amount deviation
df_behave['amount_vs_user_avg'] = df_behave['amount'] / df_behave['avg_amount_user']
df_behave['amount_zscore'] = df_behave.groupby('user_id')['amount'].transform(
    lambda x: (x - x.mean()) / x.std()
)

# Time since last transaction
# df_behave['time_since_last_txn'] = df_behave.groupby('user_id')['transaction_time'].diff().dt.total_seconds() / 3600

df_behave.head()
```

## 3- Geographical Features:

```python
df_Geo = df_behave.copy()
df_Geo['country_mismatch'] = (df_Geo['country'] != df_Geo['bin_country']).astype(int)

# Distance risk categories
df_Geo['distance_risk'] = pd.cut(df_Geo['shipping_distance_km'], 
                              bins=[0, 100, 500, 2000, 10000],
                              labels=['local', 'regional', 'cross_country', 'international'])

# High-risk shipping
df_Geo['extreme_distance'] = (df_Geo['shipping_distance_km'] > 3000).astype(int)

df_Geo.head()
```

```python
print(df_Geo.isnull().sum())
df_Geo = df_Geo.dropna(axis=0)
```

## 4- Categorical Encoding:

```python
print(df_Geo.columns.tolist())

cat_dtype_cols = df_Geo.select_dtypes(include=["object", "category"]).columns.tolist()
print("Likely categorical by dtype:", cat_dtype_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

df_cat = df_Geo.copy()

# For tree-based models (XGBoost, Random Forest)
# categorical_cols = ['country', 'bin_country', 'channel', 'merchant_category']
le = LabelEncoder()
for col in cat_dtype_cols:
    df_cat[f'{col}_encoded'] = le.fit_transform(df_cat[col])

# For neural networks, consider one-hot encoding
df_encoded = pd.get_dummies(df_cat, columns=cat_dtype_cols, drop_first=True)

# df_encoded[['bin_country_TR','bin_country_US']] = df_encoded[['bin_country_TR','bin_country_US']].astype(int)
df_encoded.head()
```

```python
import matplotlib.pyplot as plt
num_cols = df_encoded.select_dtypes(include=[np.number]).columns.tolist()
corr = df_encoded[num_cols].corr()

fig = plt.figure(figsize=(8,6))
plt.imshow(corr, aspect='auto')
plt.colorbar()
plt.xticks(range(len(num_cols)), num_cols, rotation=90)
plt.yticks(range(len(num_cols)), num_cols)
plt.title('Correlation Matrix (numeric features)')
plt.tight_layout()
plt.show()

corr.loc[['is_fraud'], :].T.sort_values('is_fraud', ascending=False).head(10)
```

```python
from sklearn.model_selection import train_test_split

drop_cols = ['is_fraud','transaction_id','transaction_time','user_id']
X = df_encoded.drop(columns=drop_cols, errors='ignore').select_dtypes(include=[np.number]).copy()
y = df_encoded['is_fraud'].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Train size:", X_train.shape, " Test size:", X_test.shape)
print("Train fraud rate:", y_train.mean(), " Test fraud rate:", y_test.mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize categorical distributions
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
categorical_cols = ['country', 'channel', 'merchant_category', 'bin_country']

for idx, col in enumerate(categorical_cols):
    ax = axes[idx // 2, idx % 2]
    df[col].value_counts().plot(kind='bar', ax=ax)
    ax.set_title(f'{col} Distribution')
    ax.set_ylabel('Count')
    
plt.tight_layout()
plt.show()
```

## Feature Importance Analysis

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

feature_importance = pd.DataFrame({
    'feature': X_train.columns,
    'importance': rf.feature_importances_
}).sort_values('importance', ascending=False)

print(feature_importance)
```

# Model 1: Bagging

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score

Bagg = BaggingClassifier(n_estimators=50, random_state=42)
Bagg.fit(X_train, y_train) 
y_pred_bagg = Bagg.predict(X_test)
y_proba_bagg = Bagg.predict_proba(X_test)[:,1]

acc_bagg = accuracy_score(y_test, y_pred_bagg)
auc_bagg = roc_auc_score(y_test, y_proba_bagg)
print(f"Accuracy: {acc_bagg:.4f}, AUC: {auc_bagg:.4f}")
print(classification_report(y_test, y_pred_bagg, digits=4))
```

# Model 2: Boosting
XGboost is used for this model

```python
import xgboost as xgb

Boost = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
Boost.fit(X_train, y_train) 
y_pred_Boost = Boost.predict(X_test)
y_proba_Boost = Boost.predict_proba(X_test)[:,1]

acc_Boost = accuracy_score(y_test, y_pred_Boost)
auc_Boost = roc_auc_score(y_test, y_proba_Boost)
print(f"Accuracy: {acc_Boost:.4f}, AUC: {auc_Boost:.4f}")
print(classification_report(y_test, y_pred_Boost, digits=4))
```

# Model 3: KNN

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn =  KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
y_proba_knn = knn.predict_proba(X_test_scaled)[:,1]

acc_knn = accuracy_score(y_test, y_pred_knn)
auc_knn = roc_auc_score(y_test, y_proba_knn)
print(f"Accuracy: {acc_knn:.4f}, AUC: {auc_knn:.4f}")
print(classification_report(y_test, y_pred_knn, digits=4))
```

# Model 4: Support Vector Machine

```python
from sklearn.svm import SVC

svc = SVC(probability=True, kernel='rbf', random_state=42)
svc.fit(X_train, y_train)

y_pred_svc = svc.predict(X_test)
y_proba_svc = svc.predict_proba(X_test)[:,1]

acc_svc = accuracy_score(y_test, y_pred_svc)
auc_svc = roc_auc_score(y_test, y_proba_svc)
print(f"Accuracy: {acc_svc:.4f}, AUC: {auc_svc:.4f}")
print(classification_report(y_test, y_pred_svc, digits=4))
```

# Model 5: Neural Network

```python
from sklearn.neural_network import MLPClassifier

nn = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42)
nn.fit(X_train, y_train)

y_pred_nn = nn.predict(X_test)
y_proba_nn = nn.predict_proba(X_test)[:,1]

acc_nn = accuracy_score(y_test, y_pred_nn)
auc_nn = roc_auc_score(y_test, y_proba_nn)
print(f"Accuracy: {acc_nn:.4f}, AUC: {auc_nn:.4f}")
print(classification_report(y_test, y_pred_nn, digits=4))
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define models
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "SVM": SVC(probability=True, kernel='rbf', random_state=42),
    "NeuralNet": MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42),
    "Bagging": BaggingClassifier(n_estimators=50, random_state=42),
    "XGBoost": xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

# Train & evaluate
results = {}
for name, model in models.items():
    print(f"\nTraining {name} ...")
    if name in ["KNN", "SVM", "NeuralNet"]:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_proba = model.predict_proba(X_test_scaled)[:,1]
    else:
        model.fit(X_train, y_train)  # for tree/ensemble models you might skip scaling
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:,1]
    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)
    print(f"{name} — Accuracy: {acc:.4f}, AUC: {auc:.4f}")
    print(classification_report(y_test, y_pred, digits=4))
    results[name] = {"accuracy": acc, "auc": auc}

# Summary
print("\nSummary of results:")
for name, metrics in results.items():
    print(f"{name}: Accuracy = {metrics['accuracy']:.4f}, AUC = {metrics['auc']:.4f}")
```
